### Find nearest polygon asset to points

```python
def find_nearest(point, buildings):
    distances = buildings.geometry.distance(point.geometry)
    min_dist = None
    nearest_id = None
    for building_id, distance in zip(buildings.id, distances):
        if min_dist is None or distance < min_dist:
            min_dist = distance
            nearest_id = building_id
    return nearest_id
```

```python
 # find nearest building to each social point
nearest_ids = []
for point in social_point_all.itertuples():
    nearest_id = find_nearest(point,buildings)
    nearest_ids.append(nearest_id)

social_point_all['nearest_building_id'] = nearest_ids
```

```python
 # add geometry from building to social point
social_nearest_polygon = social_point_all.merge(buildings, left_on='nearest_building_id', right_on='id')
social_nearest_polygon = social_nearest_polygon[[
    'id_soc_poi',
    'nearest_building_id',
    'name',
    'type',
    'geometry_y'
]]


social_nearest_polygon = social_nearest_polygon.rename(columns ={
    'geometry_y': 'geometry'})
```
