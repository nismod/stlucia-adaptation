```python
def load_sector(data_folder, sector_id):  
    sector_path = os.path.join(
        data_folder, 'data', 'buildings', '{}.shp'.format(sector_id))
    sectors = geopandas.read_file(sector_path)
    #only use valid geometries 
    sectors = sectors.loc[sectors.is_valid]
    sectors.crs = {'init' :'epsg:2006'}

    return sectors
```

```python
#for social and economic infrastructure assets 
#combine hazard exposure if two/three/four hazards are exposed

#create new column with combined hazard

#if sum of the two columns is equal or larger than 1 -> add a 1 
```

```python
landuse_area_exp = pd.read_csv(os.path.join(
    data_folder, 'results', 'landuse_hazards_area_exp.csv'))
landuse_area_exp = landuse_area_exp.drop(columns='Unnamed: 0')
```

```python
#Import and preprocess social infrastructure
social_haz_count= gpd.read_file(os.path.join(
    data_folder, 'results', 'social_hazards_capacity_exp.csv'))
social_haz_count = social_haz_count.drop(columns=['field_1'])
social_haz_count = social_haz_count.iloc[:, :-1]
social_haz_count
```

```python
#Import and preprocess economic infrastructure
econ_haz_cap = gpd.read_file(os.path.join(
    data_folder, 'results', 'econ_hazards_capacity.csv'))
econ_haz_cap = econ_haz_cap.iloc[:, :-1]
econ_haz_cap
```

```python
#combine economic and social infrastructure
econ_social = pd.concat([econ_haz_cap, social_haz_count], sort=False)
econ_social
```

```python
econ_social.info()
```

```python
hazard_ids =['1m_sea-level', '4m_storm-surge', 'flashflooding', 'landslide_susceptibility']
econ_social[hazard_ids] = econ_social[hazard_ids].apply(pd.to_numeric)
econ_social
```

```python
# Merge information for each asset - grouping by 
# -> If sector_id is the same, add the columns for each of the hazards
econ_social = econ_social.groupby([
    'sector', 'sector_id', 'admin_name', 'capacity', 'unit', 'geometry'
    ])[
    hazard_ids
    ].sum()
econ_social
```

```python
#  if hazard1 >, add capacity hazard1; else (if hazard2 >0, add capacity hazard2); else: 0  )
def two_multi_hazards(row):
    if row['4m_storm-surge'] > 0:
        val = row['4m_storm-surge']
    elif row['flashflooding'] > 0:
        val = row['flashflooding']
    else:
        val = 0 
    return val

def three_multi_hazards(row):
    if row['4m_storm-surge'] > 0:
        val = row['4m_storm-surge']
    elif row['flashflooding'] > 0:
        val = row['flashflooding']
    elif row['landslide_susceptibility'] > 0:
        val = row['landslide_susceptibility']
    else:
        val = 0 
    return val
```

```python
#create a new hazard with combined hazard 
econ_social['storm_flash'] = econ_social.apply(two_multi_hazards, axis=1)
econ_social['storm_flash_landslide']= econ_social.apply(three_multi_hazards, axis=1)
```
